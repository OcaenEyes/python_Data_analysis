```python
import pandas as pd
import numpy as np
import plotly.offline as plyoff
import plotly.graph_objs as plygo
```

```python
pyplot = plyoff.iplot
```

### 

```python
GDP = pd.read_csv("../datasets/Chinas GDP in Province Zh.csv",index_col=0)
```

```python
GDP_new = GDP.T.stack().reset_index()
```

```python
GDP_new.rename(columns={"level_0":"","level_1":"",0:"GDP"},inplace=True)
```

```python
GDP_new
```

```python
sec_house = pd.read_csv("../datasets/house.csv")
```

```python
sec_house
```

```python
sec_house_qu_price=sec_house.groupby("")[["",""]].mean().reset_index()
```

```python
sec_house_qu_price[""] = sec_house_qu_price[""] /10000
```

#### 

```python
trace0 = plygo.Scatter(x=sec_house_qu_price[""],y=sec_house_qu_price[""],mode="markers",name="")
trace1 = plygo.Scatter(x=sec_house_qu_price[""],y=sec_house_qu_price[""],mode="markers+lines",line=dict(),name="")
```

```python
data=[trace0,trace1]
```

```python
pyplot(data)
```

```python
layout = plygo.Layout(title="",
                      xaxis=dict(title=""),
                      yaxis=dict(title=""),
                      legend=dict(font=dict()),
                     )
```

```python
fig = plygo.Figure(data=data,layout=layout)
```

```python
pyplot(fig)
```

```python
trace3= plygo.Bar(x=GDP_new[GDP_new[""]==""][""],y=GDP_new[GDP_new[""]==""]["GDP"],name="")
trace4= plygo.Bar(x=GDP_new[GDP_new[""]==""][""],y=GDP_new[GDP_new[""]==""]["GDP"],name="")
trace5= plygo.Bar(x=GDP_new[GDP_new[""]==""][""],y=GDP_new[GDP_new[""]==""]["GDP"],name="")
trace6= plygo.Bar(x=GDP_new[GDP_new[""]==""][""],y=GDP_new[GDP_new[""]==""]["GDP"],name="")
```

```python
layout = plygo.Layout(title="GDP",xaxis=dict(title=""),yaxis=dict(title="GDP"))
```

```python
data = [trace3,trace4,trace5,trace6]
```

```python
fig = plygo.Figure(data=data,layout=layout)
```

```python
pyplot(fig)
```

```python
# 
layout = plygo.Layout(title="GDP",xaxis=dict(title=""),yaxis=dict(title="GDP"),barmode="stack")
data = [trace3,trace4,trace5,trace6]
fig = plygo.Figure(data=data,layout=layout)
pyplot(fig)
```
